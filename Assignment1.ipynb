Task 1 : Write a python program to check if a number is even or odd .
The program should take integer input from the user and print the whether the number is even or odd .
Example:if the user inputs 4 ,the output should be "4 is even"




```python
n=int(input('enter a number'))
if n%2==0 :
  print(f'{n} is even')
else:
  print(f'{n} is odd')
```

# Matplotlib

```python
import matplotlib.pyplot as plt
```

Task 1 :Create a linePlot using Matplotlib .
Plot the function y=x^2 for x values ranging -10 to 10.Label the axes and give the plot a title

```python
import math
x=list(range(-10,11))
y=[]
for i in x:
  y.append(math.pow(i,2))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y=x^2')
plt.show()
```

Task 2 :Create a bar chart using Matplotlib .
Use a dataset of your choice(e.g.,sales data,number of students in different classes)and plot a bar chart.Label the bars and the axes.

```python
f={'class':['6th','7th','8th','9th','10th'],'no of std in each class':[30,25,38,40,42]}
pd.DataFrame(f)
```

```python
x=f.get('class')
y=f.get('no of std in each class')
def addlabels(x,y):
  for i in range(len(x)):
    plt.text(x[i],y[i],y[i],ha='center')
plt.bar(x,y)
addlabels(x,y)
plt.xlabel('classes')
plt.ylabel('no of students in each class')
plt.title('Number of students in different classes')
plt.show()
```
